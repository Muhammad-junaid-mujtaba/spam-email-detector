```
data = '/content/email.csv'

corrected_lines = []


# Read data from the file and check for missing tabs
with open(data, 'r', encoding='utf-8') as file: # The data variable now holds the correct file path
    for i, line in enumerate(file):
        if '\t' in line:
            corrected_lines.append(line.strip())
        else:
            # Add a placeholder or fix the line if possible
            corrected_line = line.strip() + '\tPLACEHOLDER_RESPONSE'
            corrected_lines.append(corrected_line)
            print(f"Line {i + 1} fixed: {corrected_line}")

# Save the corrected lines back to the file
with open(data, 'w', encoding='utf-8') as file:
    for line in corrected_lines:
        file.write(line + '\n')

print("All lines are properly formatted with a tab character.")
```

```
inputs = []
responses = []

with open(data, 'r', encoding='utf-8') as file:
    for line in file:

        if line.startswith('ham,'):

            input_text = line[len('ham,'):].strip()
        elif line.startswith('spam,'):

            response_text = line[len('spam,'):].strip()

            inputs.append(input_text)
            responses.append(response_text)

if len(inputs) != len(responses):
    raise ValueError("Mismatch between number of questions and answers")


print(inputs)
print(responses)

token = Tokenizer()
token.fit_on_texts(inputs + responses)

input_sequences = token.texts_to_sequences(inputs)
response_sequences = token.texts_to_sequences(responses)

max_length = 100
input_sequences = pad_sequences(input_sequences, maxlen=max_length, padding='post')
response_sequences = pad_sequences(response_sequences, maxlen=max_length, padding='post')

input_sequences = np.array(input_sequences)
response_sequences = np.array(response_sequences)

le = LabelEncoder()
response_sequences = le.fit_transform(responses).reshape(-1, 1)

# Print some debug information
vac = len(token.word_index)
print(f"Number of unique words: {vac}")
output_length = len(le.classes_)
print(f"Number of unique response classes: {output_length}")
vocab_size = len(token.word_index) + 1
embedding_dim = 300
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, i in token.word_index.items():
    if word in wv:
        embedding_matrix[i] = wv[word]
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Embedding, LSTM, RepeatVector, Dense, Dropout, Flatten
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np

# Define and compile the correct model architecture
model = Sequential([
    Input(shape=(max_length,)),  # Define input shape
    Embedding(vocab_size, embedding_dim, weights=[embedding_matrix], trainable=False, input_length=max_length),  # Use pre-trained embeddings
    LSTM(128, return_sequences=False),  # LSTM layer
    Dropout(0.5),  # Dropout to prevent overfitting
    Dense(64, activation='relu'),  # Dense hidden layer
    Dropout(0.7),  # Dropout layer
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Summary to confirm correct setup
model.summary()

# Define early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Fit the model to your data
history = model.fit(
    input_sequences,
    response_sequences,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```
def predict_email(text):
    seq = token.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=max_length, padding='post')
    prediction = model.predict(padded)[0][0]
    if prediction > 0.5:
        print("The email is classified as: Spam")
    else:
        print("The email is classified as: Ham")

# Example usage
email_text = input("Enter the email text to check if it's spam or ham: ")
predict_email(email_text)
```

```
# Load pre-trained Word2Vec model from Gensim
from gensim.models import KeyedVectors
import gensim.downloader as api

# Load the pre-trained Word2Vec model
wv = api.load('word2vec-google-news-300')

# Function to get the average word vector for a given text
def get_average_word_vector(text, model):
    words = text.split()
    word_vectors = [model[word] for word in words if word in model]
    if word_vectors:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

# Convert texts to feature vectors
X_train_vectors = np.array([get_average_word_vector(text, wv) for text in X_train])
X_test_vectors = np.array([get_average_word_vector(text, wv) for text in X_test])
```

```
# Train a classifier
classifier = LogisticRegression()
classifier.fit(X_train_vectors, y_train)

# Evaluate the classifier
accuracy = classifier.score(X_test_vectors, y_test)
print(f"Model Accuracy: {accuracy:.2f}")

def predict_email(text):
    vector = get_average_word_vector(text, wv).reshape(1, -1)
    prediction = classifier.predict(vector)[0]
    label = le.inverse_transform([prediction])[0]
    return label


email_text = input("Enter the email text to check if it's spam or ham: ")
result = predict_email(email_text)
print(f"The email is classified as: {result}")
```

```
data="/content/email.csv"
df = pd.read_csv(data, header=None, names=['label', 'text'], quotechar='"', skipinitialspace=True)
inputs=df['text'].tolist()
reponse=df['label'].tolist()
le=LabelEncoder()
encoded_labels=le.fit_transform(reponse)
X_train, X_test, y_train, y_test = train_test_split(inputs, encoded_labels, test_size=0.2, random_state=42)
```

```
from gensim.models import KeyedVectors
import gensim.downloader as api

# Load the pre-trained Word2Vec model

def word(text,model):
  words=text.split()
  vec=[model[word]for word in words if word in model]
  if vec:
    return np.mean(vec,axis=0)
  else:

    return np.zeros(model.vector_size)

train=np.array([word(text,wv)for text in X_train])
test=np.array([word(text,wv)for text in X_test])
mode=LogisticRegression()
#Corrected variable names to match the training data
mode.fit(train,y_train)
accuracy=mode.score(test,y_test)
print(f"Model Accuracy: {accuracy:.2f}")
def get_result(text):
  vector=word(text,wv).reshape(1,-1)
  prediction=mode.predict(vector)[0]
  label=le.inverse_transform([prediction])[0]
  return label
email_text = input("Enter the email text to check if it's spam or ham: ")
result = get_result(email_text)
print(f"The email is classified as: {result}")
```
